   #    Cancer Detection Using HOG and LBP Features with KNN

# Part 1: Setup and Imports
In this part, we'll import the necessary libraries needed for the dataset processing, feature extraction, model training, and evaluation.


```python
# Importing necessary libraries
import os
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.feature import hog, local_binary_pattern
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import random
```

# Part 2: Load Dataset
In this part, we'll load the dataset and check whether the images and labels are available. We'll also take a random sample of 2500 images to work with.


```python
# Set your dataset path
dataset_path = "C:\\Users\\Admin\\Dataset cancer detection" 

# Check if dataset exists
if os.path.exists(dataset_path):
    print("✅ Dataset found!")
else:
    print("❌ Dataset not found! Check the path.")

# Load labels and images
train_dir = os.path.join(dataset_path, "train")
labels_path = os.path.join(dataset_path, "train_labels.csv")

# Load the CSV with labels
df = pd.read_csv(labels_path)

# Random sample of 10000 rows
df_sample = df.sample(n=10000, random_state=42).reset_index(drop=True)

images = []
labels = []

for idx, label in zip(df_sample['id'], df_sample['label']):
    img_path = os.path.join(train_dir, f"{idx}.tif")
    if os.path.exists(img_path):
        img = imread(img_path, as_gray=True)
        images.append(img)
        labels.append(label)

images = np.array(images)
labels = np.array(labels)

# Check shapes of loaded data
print("Images shape:", images.shape)
print("Labels shape:", labels.shape)
```

# Part 3: Feature Extraction (HOG and LBP)
In this part, we'll extract features from the images using Histogram of Oriented Gradients (HOG) and Local Binary Pattern (LBP) methods.

```python
X_hog = []
X_lbp = []

# HOG Parameters
hog_pixels_per_cell = (8, 8)

# LBP Parameters
lbp_radius = 1
lbp_n_points = 8 * lbp_radius

# Extract HOG and LBP features for each image
for img in images:
    # HOG
    hog_feat = hog(img, pixels_per_cell=hog_pixels_per_cell, cells_per_block=(2, 2), feature_vector=True)
    X_hog.append(hog_feat)

    # LBP
    lbp = local_binary_pattern(img, P=lbp_n_points, R=lbp_radius, method='uniform')
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, lbp_n_points + 3), range=(0, lbp_n_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6)
    X_lbp.append(hist)

X_hog = np.array(X_hog)
X_lbp = np.array(X_lbp)

print("HOG feature shape:", X_hog.shape)
print("LBP feature shape:", X_lbp.shape)
```

# Part 4: Train/Test Split
Here, we split the dataset into training and testing sets for both HOG and LBP features.

```python
X_train_hog, X_test_hog, y_train, y_test = train_test_split(X_hog, labels, test_size=0.2, random_state=42)
X_train_lbp, X_test_lbp, _, _ = train_test_split(X_lbp, labels, test_size=0.2, random_state=42)
```

# Part 5: Training KNN Classifiers on HOG and LBP
In this part, we train two KNN classifiers: one using HOG features and the other using LBP features.

```python
knn_hog = KNeighborsClassifier(n_neighbors=3)
knn_lbp = KNeighborsClassifier(n_neighbors=3)

# Train KNN for HOG features
knn_hog.fit(X_train_hog, y_train)

# Train KNN for LBP features
knn_lbp.fit(X_train_lbp, y_train)

# Make predictions
pred_hog = knn_hog.predict(X_test_hog)
pred_lbp = knn_lbp.predict(X_test_lbp)

# Print accuracy
print("HOG Accuracy:", accuracy_score(y_test, pred_hog))
print("LBP Accuracy:", accuracy_score(y_test, pred_lbp))
```

# Part 6: Fusion of Results (Weighted Voting)
Now, we combine the predictions from the HOG and LBP classifiers using a weighted voting strategy to improve the overall accuracy.

```python
hog_probs = knn_hog.predict_proba(X_test_hog)
lbp_probs = knn_lbp.predict_proba(X_test_lbp)

# Weights set based on normalized model accuracies:
# w_hog = HOG_Accuracy / (HOG_Accuracy + LBP_Accuracy) ≈ 0.4248
# w_lbp = LBP_Accuracy / (HOG_Accuracy + LBP_Accuracy) ≈ 0.5751

w_hog = 0.4248
w_lbp = 0.5751

# Weighted fusion of probabilities
fused_probs = (w_hog * hog_probs) + (w_lbp * lbp_probs)

# Convert fused probabilities to class labels
fused_preds = np.argmax(fused_probs, axis=1)

# Evaluate accuracy
fused_accuracy = accuracy_score(y_test, fused_preds)

# Print result
print("Weighted Voting Fusion Accuracy:", fused_accuracy)
```

# Part 7: Stacking Classifier
In this part, we use a stacking classifier to combine the predictions from both KNN classifiers, with a logistic regression model as the meta-model.

```python
# Part 7: Stacking Classifier
#In this part, we use a stacking classifier to combine the predictions from both KNN classifiers, with a logistic regression model as the meta-model.

# Create the base models (HOG and LBP models)
knn_hog = KNeighborsClassifier(n_neighbors=3)
knn_lbp = KNeighborsClassifier(n_neighbors=3)

# Create the meta-model (logistic regression for simplicity)
meta_model = LogisticRegression()

# Define the stacking classifier
stacking_clf = StackingClassifier(estimators=[('hog', knn_hog), ('lbp', knn_lbp)], final_estimator=meta_model)

# Train the stacking classifier
stacking_clf.fit(X_train_hog, y_train)  # You can use either feature set for training

# Predict with the stacking classifier
stacking_pred = stacking_clf.predict(X_test_hog)  # or X_test_lbp

# Evaluate accuracy
stacking_accuracy = accuracy_score(y_test, stacking_pred)

# Print result
print("Stacking Accuracy:", stacking_accuracy)
```

# Part 8: Visualize Image and Features

# Part 9: Confusion Matrix
We will now plot the confusion matrix for each of the models (HOG, LBP, Weighted Voting, and Stacking).

```python
# Function to visualize Confusion Matrix
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

# Confusion Matrix for HOG model
plot_confusion_matrix(y_test, pred_hog, "HOG Model")

# Confusion Matrix for LBP model
plot_confusion_matrix(y_test, pred_lbp, "LBP Model")

# Confusion Matrix for Weighted Voting Results
plot_confusion_matrix(y_test, fused_preds, "Weighted Voting Model")

# Confusion Matrix for Stacking Model
plot_confusion_matrix(y_test, stacking_pred, "Stacking Model")
```

# Part 10: ROC Curve Analysis
We will analyze the ROC curve for each model (HOG, LBP, Weighted Voting, and Stacking) to evaluate their performance.

```python
#Train individual KNN classifiers
knn_hog = KNeighborsClassifier(n_neighbors=3)
knn_lbp = KNeighborsClassifier(n_neighbors=3)

knn_hog.fit(X_train_hog, y_train)
knn_lbp.fit(X_train_lbp, y_train)

#Predict probabilities for weighted fusion
hog_probs = knn_hog.predict_proba(X_test_hog)[:, 1]
lbp_probs = knn_lbp.predict_proba(X_test_lbp)[:, 1]

#Weighted Average Fusion (you can adjust weights as needed)
w1, w2 = 0.4248, 0.5751
weighted_probs = w1 * hog_probs + w2 * lbp_probs

#Create fused feature sets for stacking
X_train_fused = np.hstack((X_train_hog, X_train_lbp))
X_test_fused = np.hstack((X_test_hog, X_test_lbp))

#Stacking Classifier for fusion
stacking_clf = StackingClassifier(
    estimators=[('hog', knn_hog), ('lbp', knn_lbp)],
    final_estimator=LogisticRegression()
)
stacking_clf.fit(X_train_fused, y_train)
stacking_probs = stacking_clf.predict_proba(X_test_fused)[:, 1]

#ROC Curves
fpr_hog, tpr_hog, _ = roc_curve(y_test, hog_probs)
fpr_lbp, tpr_lbp, _ = roc_curve(y_test, lbp_probs)
fpr_weighted, tpr_weighted, _ = roc_curve(y_test, weighted_probs)
fpr_stacking, tpr_stacking, _ = roc_curve(y_test, stacking_probs)

roc_auc_hog = auc(fpr_hog, tpr_hog)
roc_auc_lbp = auc(fpr_lbp, tpr_lbp)
roc_auc_weighted = auc(fpr_weighted, tpr_weighted)
roc_auc_stacking = auc(fpr_stacking, tpr_stacking)

#Plot ROC Curves
plt.figure(figsize=(10, 7))
plt.plot(fpr_hog, tpr_hog, label=f'HOG KNN (AUC = {roc_auc_hog:.2f})')
plt.plot(fpr_lbp, tpr_lbp, label=f'LBP KNN (AUC = {roc_auc_lbp:.2f})')
plt.plot(fpr_weighted, tpr_weighted, label=f'Weighted Fusion (AUC = {roc_auc_weighted:.2f})')
plt.plot(fpr_stacking, tpr_stacking, label=f'Stacking Fusion (AUC = {roc_auc_stacking:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Weighted Fusion and Stacking')
plt.legend()
plt.grid(True)
plt.show()
```
